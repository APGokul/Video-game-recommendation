```python
import pandas as pd
import numpy as np
```

```python
games = pd.read_csv('games.csv')
```

```python
games.head()
```

```python
games = games[['Title','Team', 'Genres', 'Summary']]
```

```python
games.head()
```

```python
games.shape
```

```python
games.info()
```

```python
games.isnull().sum()
```

```python
games.dropna(inplace=True)
```

```python
games.drop_duplicates(inplace=True, keep='first')
```

```python
games.duplicated().sum()
```

```python
games['Summary'].head()
```

```python
games['Summary'] = games['Summary'].apply(lambda x:x.split())
```

```python
games.head()
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i)
    return L
        
```

```python
games['Team'] = games['Team'].apply(convert)
games['Genres'] = games['Genres'].apply(convert)
```

```python
games['Team'].head()
```

```python
games['Team'] = games['Team'].apply(lambda x:[i.replace(" ","") for i in x])
games['Genres'] = games['Genres'].apply(lambda x:[i.replace(" ","") for i in x])
games['Summary'] = games['Summary'].apply(lambda x:[i.replace(" ","") for i in x])


```

```python
games.head()
```

```python
games['tags'] = games['Team'] + games['Genres'] + games['Summary']
```

```python
games.head()
```

```python
new_df = games[['Title','tags']] 
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
new_df.shape
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
new_df[new_df['Title']== 'Grand Theft Auto V'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key= lambda x:x[1])
```

```python
def recommend(movie):
    movie_index = new_df[new_df['Title']== movie].index[0]
    distances = similarity[movie_index] 
    movies_list = sorted(list(enumerate(distances)),reverse=True, key= lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].Title)
```

```python
recommend('The Witcher 3: Wild Hunt')
```
